```python
import pandas as pd
```

```python
# data_df = pd.read_csv('./아파트(매매)__실거래가_20220528222001.csv', encoding='cp949', header=15)
data_df = pd.read_csv('./연립다세대(매매)__실거래가_20220528224149.csv', encoding='cp949', header=15)

data_df
```

```python
data_df.dtypes
```

```python
data_df['거래금액'] = data_df['거래금액(만원)'].str.replace(',','').astype('int64')
```

```python
data_df['전용면적'] = data_df['전용면적(㎡)'].astype('int32')
```

```python
data_df_mean = data_df.groupby('전용면적')['거래금액'].mean()
# data_df_mean
```

```python
data_df_max = data_df.groupby('전용면적')['거래금액'].max()
data_df_median = data_df.groupby('전용면적')['거래금액'].median()
data_df_min = data_df.groupby('전용면적')['거래금액'].min()
```

```python
df_concat = pd.concat([data_df_mean, data_df_max, data_df_median,data_df_min], axis=1)
df_concat.columns = ['mean', 'max', 'median', 'min']
df_concat.reset_index(inplace=True)
df_concat
```

```python
df_concat[(df_concat['전용면적'] >= 60) & (df_concat['전용면적'] <= 70)]
```
