```python
import pandas as pd
```

```python
# data_df = pd.read_csv('./아파트(매매)__실거래가_20220528222001.csv', encoding='cp949', header=15)
data_df = pd.read_csv('./연립다세대(매매)__실거래가_20220528224149.csv', encoding='cp949', header=15)

data_df
```

```python
data_df.dtypes
```

```python
data_df['거래금액'] = data_df['거래금액(만원)'].str.replace(',','').astype('int64')
```

```python
data_df['전용면적'] = data_df['전용면적(㎡)'].astype('int32')
```

```python
data_df_groupby = data_df.groupby('전용면적')['거래금액'].mean()
data_df_groupby
```

```python
type(data_df_groupby)
```

```python
data_df_group = pd.DataFrame(data_df_groupby)
data_df_group.reset_index(inplace=True)
data_df_group[data_df_group['전용면적'] >= 60]
```
