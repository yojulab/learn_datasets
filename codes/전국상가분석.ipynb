```python
import glob

filename_list = []
for file_name in glob.glob("./*.csv"):
    filename_list.append(file_name)
filename_list
```

##read csv from files and concat DataFrammes

```python
import pandas as pd
```

```python
df_csv_list = list()
for file_name in filename_list:
  df_csv_list.append(pd.read_csv(file_name))

df_concat = pd.concat(df_csv_list)
# df_concat
```

```python
# show all rows
# pd.options.display.max_rows = df_concat.shape[0]
df_concat
```

```python
df_concat.reset_index(inplace=True)
df_concat
```

##데이터 전처리

###서브셋 만들기 : 필요한 컬럼들 선택

```python
df_concat.columns
```

```python
store_columns = ['상호명','상권업종대분류명', '상권업종중분류명', '상권업종소분류명', '시도명', '시군구명', '행정동명', '법정동명','경도', '위도']
```

```python
stores_df = df_concat[store_columns]
# stores_df.shape, stores_df.info()
stores_df.isnull().sum()
```

###업종 코드 종류 : 대,중,소

```python
stores_category_columns = ['상권업종대분류명', '상권업종중분류명', '상권업종소분류명']
stores_df[stores_category_columns].nunique()
```

```python
stores_category_df = stores_df[stores_category_columns]
stores_category_df
```

```python
stores_category_df = stores_df[stores_category_columns]
store_categories_df = stores_category_df.drop_duplicates().sort_values(stores_category_columns)
store_categories_df
```

```python
# pd.options.display.max_rows = 20 # len(store_categories_df)
store_categories_df.value_counts().reset_index()
```

##전국 상가 등록 현황

###대분류별 업종 상가 갯수

```python
stroe_category_large_series = stores_category_df['상권업종대분류명'].value_counts()
stroe_category_large_series
```

####그래프 그리기

```python
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 

fig = plt.figure(figsize=(10,5))
plt.subplot(121)
plt.bar(stroe_category_large_series.index, stroe_category_large_series.values)
plt.xticks(rotation=45)
plt.grid(axis='y', ls=':')

plt.subplot(122)
plt.pie(stroe_category_large_series, labels=stroe_category_large_series.index
        , autopct='%.1f%%', pctdistance=0.8, wedgeprops={'width':0.6})

fig.suptitle('업종별 전국 상가 현황', size=20)
fig.tight_layout()
plt.show()
```
