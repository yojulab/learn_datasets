```python
import glob

filename_list = []
for file_name in glob.glob("./*.csv"):
    filename_list.append(file_name)
filename_list
```

##read csv from files and concat DataFrammes

```python
import pandas as pd
```

```python
df_csv_list = list()
for file_name in filename_list:
  df_csv_list.append(pd.read_csv(file_name))

df_concat = pd.concat(df_csv_list)
# df_concat
```

```python
# show all rows
# pd.options.display.max_rows = df_concat.shape[0]
df_concat
```

```python
df_concat.columns
```

```python
stores_category_columns = ['상권업종대분류명', '상권업종중분류명', '상권업종소분류명']
stores_category_df = df_concat[stores_category_columns]
pd.options.display.max_rows = 20
stores_category_df
```

```python
store_categories_df = stores_category_df.drop_duplicates().sort_values(stores_category_columns)
store_categories_df
```
