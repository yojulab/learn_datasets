# 실습1. 웹 크롤링

- request.get으로 크래핑
- BeautifulSoup로 lxml 또는 html로 tag 정렬

```python
import requests
```

```python
from bs4 import BeautifulSoup
url = 'https://www.naver.com/'
r = requests.get(url)
bs2 = BeautifulSoup(r.text, 'html.parser')
rt = bs2.find('a', class_='al_favorite')
type(rt.attrs['href']), rt.get_text()
```

## 크롤링 기본 코드

```python
url = 'https://finance.naver.com/item/sise_day.nhn?code=005930&page=2'
r = requests.get(url)
r.text
```

```python
type(r.text)
```

```python
from bs4 import BeautifulSoup
```

```python
bs = BeautifulSoup(r.text, 'lxml')
bs2 = BeautifulSoup(r.text, 'html.parser')
```

```python
bs
```

```python
type(bs)
```

```python
rt = bs.find('table', class_='type2')
rt
```

```python
rt2 = rt.findAll('tr')
```

```python
len(rt2)
```

```python
rt2[3]
```

```python
for idx, x in enumerate(rt2):
    print(idx, x.text.strip())
```

```python
rt3 = rt2[2].findAll('td')
rt3
```

```python
len(rt3)
```

```python
for x in rt3:
    print(x.text.strip())
```

```python
total = [] # total에 리스트로 삽입

for idx, x in enumerate(rt2):
    inner_list = []
    x2 = x.findAll('td')
    if(len(x2) < 3):
        continue
    for cnt in x2:
        print(cnt.text.strip(), end=', ')
        inner_list.append(cnt.text.strip())
    print()
    total.append(inner_list)
```
